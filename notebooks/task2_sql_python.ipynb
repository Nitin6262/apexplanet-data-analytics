```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///../apexplanet.db")

print("Database connected successfully using SQLAlchemy!")
```

```python
query = "SELECT * FROM superstore LIMIT 10;"

df = pd.read_sql(query, engine)

df
```

## Business Question 1

### Which category generated the highest total sales?

```python
query = """
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

result = pd.read_sql(query, engine)

result
```

## Business Question 2

### Which 10 customers generated the highest total sales?

```python
query = """
SELECT "Customer Name",
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

result = pd.read_sql(query, engine)

result
```

## Business Question 6

### Which ship mode was used the most?

```python
query = """
SELECT "Ship Mode",
       COUNT(*) AS Total_Orders
FROM superstore
GROUP BY "Ship Mode"
ORDER BY Total_Orders DESC;
"""

result = pd.read_sql(query, engine)
result
```

## Business Question 8

### What is the average discount by category?

```python
query = """
SELECT Category,
       AVG(Discount) AS Average_Discount
FROM superstore
GROUP BY Category;
"""

result = pd.read_sql(query, engine)
result
```

## Business Question 9

### What is the monthly sales trend?

```python
query = """
SELECT
strftime('%m', "Order Date") AS Month,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Month
ORDER BY Month;
"""

result = pd.read_sql(query, engine)
result
```
